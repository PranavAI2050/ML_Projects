# Black Friday EDA and Feature Engineering

```python
df_train = pd.read_csv("/kaggle/input/black-friday/train.csv")
df_test  = pd.read_csv("/kaggle/input/black-friday/test.csv")
```

```python
df_train.head()
```

```python
df_test.loc[:,'Purchase'] = -1
df_combined = pd.concat([df_train,df_test], axis = 0)
```

```python
df_combined.isnull().sum()
```

```python
df_combined.info()
```

    User Id is Useless as it is a unique id given to each customer Therfore we drop it

```python
df_combined  = df.drop('User_ID', axis = 1)
```

```python
df_combined['Gender'] = df_combined.Gender.map({'F':0,'M':1})                      
```

```python
df_combined['Gender'] = pd.get_dummies(df['Gender'], drop_first = 1)
```

```python
df_combined.head()
```

```python
age = {'0-17':0, '55+':6, '26-35':2, '46-50':4, '51-55':5, '36-45':3, '18-25':1}
df_combined['Age'] = df.Age.map(age)
```

```python
df_combined.drop('City_Category',axis = 1, inplace = True)
```

```python
df_combined.isnull().sum()
```

```python
df_combined.Product_Category_2.value_counts()
```

```python
df_combined['Product_Category_2'] = df_combined.Product_Category_2.fillna(df_combined['Product_Category_2'].mode()[0])
```

```python
df_combined['Product_Category_2']
```

```python
df_combined['Product_Category_3'] = df_combined.Product_Category_3.fillna(df_combined['Product_Category_3'].mode()[0])
```

```python
Stay_years = {'2':2, '4+':4, '3':3, '1':1, '0':0}
df_combined['Stay_In_Current_City_Years'] = df_combined.Stay_In_Current_City_Years.map(Stay_years)
```
